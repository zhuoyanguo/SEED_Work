```python
import pandas as pd

#Read in data
df = pd.read_csv('/content/SEED DC SEL_Engagement Assessment Results - Winter 2024 - SEED DC Winter SEL Survey.csv')

df.head()
```

```python
# Extract the student name and delete all the "Please select your name" columns
# Identify all columns that start with "Please select your name"
name_columns = [col for col in df.columns if col.startswith("Please select your name")]

print(name_columns)
```

```python
# Create a new column "Student Name" by extracting the first non-NA value from the identified columns for each row
df['Student Name'] = df[name_columns].apply(lambda row: next(item for item in row if pd.notna(item)), axis=1)

# Display the updated dataframe to verify the new column has been added correctly
df[['Student Name'] + name_columns].head()
```

```python
# Delete all the "Please select your name" columns from the dataframe and also delete the Timestamp
df = df.drop(columns=name_columns)
df = df.drop(columns=['Timestamp'])

df.head()
```

```python


# Reorder the dataframe to make "Student Name" the first column
column_order = ['Student Name'] + [col for col in df.columns if col != 'Student Name']
df = df[column_order]

df.head()
```

```python
# Import Completeion Tracker data
df_tracker = pd.read_csv('/content/SEED DC SEL_Engagement Assessment Completion Tracker - Winter 2024 - Winter 2024 Completion Tracker.csv', skiprows=8)
df_tracker.head()
```

```python
# Setting the correct headers for the tracker dataframe
df_tracker.columns = df_tracker.iloc[0]  # Set the first row as the header
df_tracker = df_tracker.drop(0)  # Drop the first row now that headers are set

df_tracker.head()
```

```python
# Filtering the DataFrame to show only rows where 'Completed?' is FALSE
false_rows = df_tracker[df_tracker['SEL Assessment \nCompleted?'] == 'FALSE']

# Displaying the rows
print(false_rows)
```

```python
# Check with the Completion Tracker to make sure correct student id and grade level
# Merging the original dataset with the tracker dataset to update "Student ID" and "Grade Level" based on "Full Name"
# Using a left merge to ensure all records from the original dataset are retained
df_merged = pd.merge(df_tracker[['Full Name', 'ID', 'Grade Level', 'Gender']], df, left_on='Full Name', right_on='Student Name', how='left')

df_merged.head()
```

```python
# Melting the dataframe to convert question columns into row entries
df_melted = df_merged.melt(id_vars=['Full Name', 'Student ID', 'Gender', 'Grade Level', 'Race/Ethnicity'],
                                  var_name='Question', value_name='Response')

df_melted.head()
```

```python
question_category = pd.read_csv('/content/SEED DC SEL_Engagement Assessment Results - Winter 2024 - Question Categories.csv')

question_category.head()
```

```python

question_category.columns = ['Survey', 'Question', 'Category']
question_category = question_category[question_category['Survey'] != 'YAPSS']
question_category.head()
```

```python
# Add Category to df_melted
# Merging the melted dataframe with the categories dataframe to assign categories to each question
df = pd.merge(df_melted, question_category, on='Question', how='left')

# After the merge, drop Survey column
df.drop(columns=['Survey', 'Full Name'], inplace=True)

df
```

```python
# Handle the duplicate questions
df['Question'] = df['Question'].astype(str).str.replace('\.1$', '', regex=True)

df = df.drop_duplicates(subset=['Student ID', 'Question'], keep='first')

len(df)
```

```python
# Adding Numerical Response
import re
# Function to apply the logic
def score_response(row):
    # Convert response to string in case it's not
    response = str(row['Response']).lower()
    category = row['Category']

    if category == 'Mental Health':
        if re.search(r'at no time|not', response):
            return 0
        elif re.search(r'some of the time|little', response):
            return 0.8
        elif re.search(r'less than half|somewhat', response):
            return 1.6
        elif re.search(r'more than half|quite', response):
            return 2.4
        elif re.search(r'most', response):
            return 3.2
        elif re.search(r'all of the|lot', response):
            return 4
    else:
        if re.search(r'at all|almost never', response):
            return 0
        elif re.search(r'little|slightly|once', response):
            return 1
        elif re.search(r'some', response):
            return 2
        elif re.search(r'quite|often|frequently|pretty', response):
            return 3
        elif re.search(r'completely|extremely|almost al|tremendous|a lot|all of the', response):
            return 4
        elif response == 'yes':  # Converted to lowercase
            return 4
        elif response == 'no':  # Converted to lowercase
            return 0
    return ""  # Return an empty string if no condition is met

# Applying the function to each row
df['Numerical Response'] = df.apply(score_response, axis=1)
df.head()
```

```python
#Adding Category Score
# Function to apply the logic
import numpy as np

# Convert the 'G' column to numeric, coercing errors to NaN (in case there are non-numeric values)
df['Numerical Response'] = pd.to_numeric(df['Numerical Response'], errors='coerce')

def calculate_average(row, df):
    # Filter the DataFrame based on Student_ID and Category matching the current row's values
    filtered_df = df[(df['Student ID'] == row['Student ID']) & (df['Category'] == row['Category'])]

    # Calculate the average if there are any matching rows
    if not filtered_df.empty:
        return filtered_df['Numerical Response'].mean() * 10
    else:
        # Return NaN or another specified value if there are no matches
        return np.nan

# Applying the function to each row
df['Category Score'] = df.apply(calculate_average, axis=1, args=(df,))
df.head()
```

```python
# Adding Category Score

# Function to apply the logic
def check_category_score(score):
    if pd.isnull(score):  # Check if the score is NaN
        return np.nan  # Return NaN if the score is blank
    else:
        return int(score >= 28)  # Return 1 if score >= 28, else 0, as an integer


# Applying the function to the "Category Score" column
df['Category Threshold'] = df['Category Score'].apply(check_category_score)

df.head()
```
